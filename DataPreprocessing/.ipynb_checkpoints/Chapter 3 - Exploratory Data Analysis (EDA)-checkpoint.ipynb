```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import random
import scipy.stats as stats
%matplotlib inline 
sns.set()
```

```python
churn = pd.read_csv("data/churn.txt")
summary_churn = churn.describe()
churn.columns = ['State', 'Account Length', 'Area Code', 'Phone', 'Intl Plan',
       'VMail Plan', 'VMail Message', 'Day Mins', 'Day Calls', 'Day Charge',
       'Eve Mins', 'Eve Calls', 'Eve Charge', 'Night Mins', 'Night Calls',
       'Night Charge', 'Intl Mins', 'Intl Calls', 'Intl Charge',
       'CustServ Calls', 'Churn']

# We need to fix Churn flag, replace "." with "" and then convert it to bool.
churn.Churn = churn.Churn.apply(lambda x: eval(x.replace(".","")))
prop_churn = (churn.Churn.value_counts()[True] / churn.Churn.count()) * 100
print("Proportion of those who churn is {}%".format(prop_churn.round(2)))
```

```python
# H0 - There is no difference in churners between population with 'Intl Calls' and without 'Intl Calls'
# H1 - There is significant difference in churners between these two population.

# A) Point Estimate
#    1) Population sample for each dataset (Churn = True and Churn = False).
churn_true=  churn[churn.Churn.loc[:'True']]
churn_false = churn[~churn.Churn.loc[:'True']]

#Welch Two sample t-test
tscore, pvalue  = stats.ttest_ind(churn_false['Intl Calls'],churn_true['Intl Calls'] , equal_var=False)

print(tscore,pvalue)
# High score of t-test (around 3) indicates that there is difference in churners for the given populations
# p-value < .05 means that the difference observed in not fluke or by chance but statistically significant.

# B) Confidence Interval - http://sphweb.bumc.bu.edu/otlt/MPH-Modules/BS/BS704_Confidence_Intervals/BS704_Confidence_Intervals5.html
    #If we would have run this test in R, we would have got more reporting details (which are very handy)
    # Lets calculate those in python using formulla's

#   1)
mean_of_true = churn_true['Intl Calls'].mean()
mean_of_false = churn_false['Intl Calls'].mean()

#   2) 95% Confidence interval - What is lower and upper bounds of difference in means between given population 
#      that we can be 95% confident about

# Confidence interval = Point Estimate +/- Margin of Error.
# Margin of error = MOE = t95_df * S * sqrt(1/(N1) + 1/(N1))
# t.95 - t score at 95% interval for a given degree of freedom. 
#        This can either be looked in Statistics ttable or fetched from Python Stats library.
# df - degree of freedom.

N1 = len(churn_true)
N2 = len(churn_false)
df = N1 + N2 - 2 # Standard 2 - sample

const = 1/(1 + (N1 * std2**2)/(N2 * std1**2))
df =  1/ (const**2/(N1-1) +  ((1-const)**2/(N2-1))) #Welch
                         
                         
std1 = (churn_true['Intl Calls'].std())
std2 = (churn_false['Intl Calls'].std())

#S = sqrt( ((N1 - 1)*(std1)**2 + (N2 - 1)*(std2)**2) / df) 
S = np.sqrt( ((N1 - 1)*(std1)**2 + (N2 - 1)*(std2)**2) / df) # 
S1 = np.sqrt(std1**2/N1 + std2**2/N2) #Welch
t95_df = stats.t.ppf([1 - (1-.95)/2], df)

t95_df[0] = (t95_df[0])

#(1/sqrt(N1) + 1/sqrt(N1))
sq_n1_n2 = round(np.sqrt(1/N1 + 1/N2),2)

moe = t95_df * S * sq_n1_n2
moe1 = t95_df * S1 #Welch
print("S1", S1)
print("Confidence Interval1: ", mean_of_false - mean_of_true - moe1, mean_of_false - mean_of_true + moe1)
print("Params", mean_of_true, std1, N1, mean_of_false, std2,N2)
print("t95", (t95_df))
print("S", round(S,2))
print("sq_n1_n2", sq_n1_n2)
print("df:", df)
print("mean_true:", mean_of_true, " mean_false:", mean_of_false)
print("Confidence Interval: ", mean_of_false - mean_of_true - moe, mean_of_false - mean_of_true + moe)
```
