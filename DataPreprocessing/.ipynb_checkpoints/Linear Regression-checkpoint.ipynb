```python
from sklearn import datasets, linear_model
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pylab as plt

%matplotlib inline 
```

```python
iris = datasets.load_iris()

i = pd.DataFrame(data= np.c_[iris['data'], iris['target']],
                     columns= iris['feature_names'] + ['target'])
```

```python
i.info()
i.columns

X = i['petal width (cm)']
y = i['sepal length (cm)']
print(X.shape, y.shape)
```

```python
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
sns.lmplot('petal width (cm)','sepal length (cm)', data=i)
```

```python
sns.residplot('petal width (cm)','sepal length (cm)', data=i)
```

```python
reg = linear_model.LinearRegression()
X = X.reshape(-1,1)
y = y.reshape(-1,1)
reg.fit(X,y)
```

```python
# Other way to run linear regression (If F-statistic is required.)
X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)
reg.score(X_test, y_test)
```

```python
X2_train = sm.add_constant(X_train)

est = sm.OLS(y_train, X2_train)
est2 = est.fit()
X2_test = sm.add_constant(X_test)
y_predict = est2.predict(X2_test)

y_predict.reshape(-1,1)
print(y_predict)
print(y_test.reshape(1,-1)[0])
```

```python
lasso = Lasso(alpha = 0.1)
X = i.drop('sepal width (cm)', axis = 1)

names = X.columns
y = i[['sepal width (cm)']]
lasso_coeff = lasso.fit(X, y).coef_
_ = plt.plot(range(len(names)), lasso_coeff)
plt.xticks(range(len(names)), names)
plt.ylabel('Coeff')
```

```python
xa = i[['petal length (cm)','petal width (cm)']]
p = PolynomialFeatures( interaction_only = True, include_bias = False)
X = p.fit_transform(xa)

y = i[['sepal length (cm)']]
reg = linear_model.LinearRegression()

reg.fit(X,y)
print("R2: ", str(reg.score(X,y))) #R2
print ("Intercept: " , str(reg.intercept_)) #Intercept
print("Coeffs: ", str(reg.coef_))
```
